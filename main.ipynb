```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**3
y
```

```python
plt.plot(x,y)
plt.title("X-Y Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,x**2)

plt.subplot(2,2,2)
plt.plot(y,y**2)

plt.subplot(2,2,3)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.plot(y,x)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1]) 
axis1.plot(x,y)

axis2 = fig.add_axes([0.2,0.2,0.3,0.3])  #0.3,0.3 = Height/Width , 0.1,0.1 = Distance from x/y axis
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize = (20,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,1,1]) 
axis1.plot(x,y)

axis2 = fig.add_axes([0.2,0.2,0.3,0.3])  #0.3,0.3 = Height/Width , 0.1,0.1 = Distance from x/y axis
axis2.plot(x,y)
```
